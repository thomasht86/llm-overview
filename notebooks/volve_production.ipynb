```python
# Histogram of oil volumes
fig = px.histogram(df, x="BORE_OIL_VOL")
fig.show()

# Histogram of gas volumes
fig = px.histogram(df, x="BORE_GAS_VOL")
fig.show()

# Histogram of water volumes
fig = px.histogram(df, x="BORE_WAT_VOL")
fig.show()
```
